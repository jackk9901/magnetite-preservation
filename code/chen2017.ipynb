```python
# Jack Kreisler, Dartmouth College, May 2020

# import and read stuff
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("data/chen2017_MD12-3432_Ms-age-TOC-IW.xlsx")
df
```

```python
# select columns
depth = df["Depth [m]"]
ms = df["Ms [nA/m**2]"]
mrs = df["Mrs [nA/m**2]"]
```

```python
# plot and show
plt.figure()
plt.plot(ms, depth, "ro")
plt.xlabel("Ms [nA/m**2]")
plt.ylabel("Depth [m]")
plt.figure()
plt.plot(ms, mrs, "bo")
plt.xlabel("Ms [nA/m**2]")
plt.ylabel("Mrs [nA/m**2]")
#plt.show()
#plt.savefig("test-save.pdf")

```
